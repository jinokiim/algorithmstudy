```python
N = int(input())

board = [list(map(int, input())) for _ in range(N)]
```

```python
board
```

```python
check = [[False] * N for _ in range(N)]
```

```python
check
```

```python
result = []
each = 0
dy = [0,1,0,-1]
dx = [1,0,-1,0]
```

```python
def dfs(y, x):
    for k in range(4):
        ny = y + dy[k]
        nx = x + dx[k]
        if 0<=ny<N and 0<=nx<N:
            if board[ny][nx] == 1 and check[ny][nx] == False:
                check[ny][nx] = True
                dfs(ny,nx)


for j in range(N):
    for i in range(N):
        if board[j][i] == 1 and check[j][i] == False:
            # 방문 체크 표시
            check[j][i] = True
            # dfs로 크기 구하기
            each = 0
            dfs(j,i)

            # 크기를 결과 리스트에 넣기
            result.append(each)

result.sort()
print(len(result))
for i in result:
    print(i)
```
