```python
from collections import deque

n, m = map(int, input().split())

board = [list(map(int, input().split())) for _ in range(n)]
check = [[False] * m for _ in range(n)]
```

```python
count = 0
max_value = 0
```

```python
dy = [0,1,0,-1]
dx = [1,0,-1,0]

def bfs(y, x):
    result = 1
    q = deque()
    q.append((y, x))
    while q:
        ey, ex = q.popleft()
        for k in range(4):
            ny = ey + dy[k]
            nx = ex + dx[k]
            
            if 0<=ny<n and 0<=nx<m:
                if board[ny][nx] == 1 and check[ny][nx] == False:
                    result += 1
                    check[ny][nx] = True
                    q.append((ny,nx))
    return result
                    
    

for j in range(n):
    for i in range(m):
        if board[j][i] == 1 and check[j][i] == False:
            check[j][i] = True
            count += 1
            max_value = max(max_value, bfs(j,i))
            
print(count)
print(max_value)
```
