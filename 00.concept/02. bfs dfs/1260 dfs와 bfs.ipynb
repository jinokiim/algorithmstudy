```python
'''
4 5 1
1 2
1 3
1 4
2 4
3 4
'''
from collections import deque
graph = deque()

N, M, V = map(int, input().split())

graph = [[] for _ in range(N+1)]
```

```python
graph
```

```python
visited = [False] * (N + 1)
```

```python
for i in range(M):
    a, b = map(int, input().split())
    graph[a].append(b)
    graph[b].append(a)
    
graph
```

```python
for i in range(len(graph)):
    graph[i].sort()
    
graph
```

```python
def dfs(n):
    visited[n] = True
    print(n, end=' ')
    for i in graph[n]:
        if not visited[i]:
            dfs(i)
```

```python
def bfs(n):
    queue = deque([n])
    visited[n] = True
    while queue:
        v = queue.popleft()
        print(v, end=' ')
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
```

```python
dfs(V)
```
